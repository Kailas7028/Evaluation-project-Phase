```python
#split features and label and drop features with high vif score to reduce mutlicollinearity
X=df.drop(columns=['W',],axis=1)
y=df['W']
```

```python
#normaliaze scale
scal=StandardScaler()
x_scal=scal.fit_transform(X)
```

```python
#feature selection
selectkbest=SelectKBest(score_func=f_regression,k=11)
x_best=selectkbest.fit_transform(x_scal,y)
```

```python
#find best random state for model
maxscr=0
ind=0
for i in range(100):
  x_train,x_test,y_train,y_test=train_test_split(x_best,y,test_size=0.20,random_state=i)
  lr=LinearRegression()
  lr.fit(x_train,y_train)
  y_pred=lr.predict(x_test)
  scr=r2_score(y_test,y_pred)
  if scr > maxscr:
    maxscr=scr
    ind=i
```

```python
print(f'max score {maxscr} at index {ind}')
```

**Model Building**

```python
#spliting data for training and testing
x_train,x_test,y_train,y_test=train_test_split(x_best,y,test_size=0.20,random_state=ind)
```

```python
def train_model(x_tr,x_te,y_tr,y_te,alg):
  alg.fit(x_tr,y_tr)
  print('-'*25,'Training Score','-'*25)
  y_pred_tr=alg.predict(x_tr)
  tr_scr=r2_score(y_tr,y_pred_tr)
  print(tr_scr*100)
  print('-'*25,'Testing Score','-'*25)
  y_pred_te=alg.predict(x_te)
  te_scr=r2_score(y_te,y_pred_te)
  print(te_scr*100)
  return alg
```

```python
#building linearregressor
lr=train_model(x_train,x_test,y_train,y_test,LinearRegression())
```

```python
#building kneighbors
kn=train_model(x_train,x_test,y_train,y_test,KNeighborsRegressor())
```

```python
#building SVR
svr=train_model(x_train,x_test,y_train,y_test,SVR())
```

```python
#building Decisiontreregressor
dt=train_model(x_train,x_test,y_train,y_test,DecisionTreeRegressor())
```

```python
#building Randomforestregressor
rf=train_model(x_train,x_test,y_train,y_test,RandomForestRegressor())
```

```python
#building AdaboostRegressor
adr=train_model(x_train,x_test,y_train,y_test,AdaBoostRegressor())
```

```python
#building GradientBoostingRegressor
gbr=train_model(x_train,x_test,y_train,y_test,GradientBoostingRegressor())
```
